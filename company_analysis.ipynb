# <center> **Building company analysis for stock pick** </center>

This notebook has as purpose the construction of an analysis based in fundamentalist concepts.

We are going to use the website www.fundamentus.com.br to retrieve companies data

```python
import pandas as pd
import numpy as np
```

```python
# Using pandas to read company data table retrieved from http://www.fundamentus.com.br/resultado.php

url = 'http://www.fundamentus.com.br/resultado.php'
dataframe = pd.read_html(url, decimal=',', thousands='.')[0]
```

```python
dataframe
```

```python
# First of all, to make comparisons, we need to transform all the data that aren't numeric to numeric
# Those columns are

0	Div.Yield
1	Mrg Ebit
2	Mrg. Líq.
3	ROIC
4	ROE
5	Cresc. Rec.5a


str_cols = dataframe.select_dtypes(['object']).columns
str_cols = str_cols[1:]
def parse_string(data):
    data[str_cols] = data[str_cols].replace('\.', '', regex=True)
    data[str_cols] = data[str_cols].replace(',', '.', regex=True)
    data[str_cols] = data[str_cols].apply(lambda x: x.str.strip('%')).astype(float)/100
```

```python
%%time 
parse_string(dataframe)
```

```python
dataframe
```

Here we start filtering actions
All these filters are arbitraries using personal opinions - You can modify the filters with your own personal strategy

The filters here are:
- Companies with Average daily volume traded in the last 2 months greater than 500.000
- Company with Dividend Yield greater than 2%
- Positive P/L (Company with profits)

```python
filtered_df = dataframe[(dataframe['Liq.2meses'] > 500000) & (dataframe['Div.Yield'] > 0.02)]
filtered_df = filtered_df[filtered_df['P/L'] > 0].sort_values(by=['P/L'], ascending = True)
```

Now, we are going to calculate two more filters and create columns with the calculated values
- Graham Formula: Better known as intrinsic value of a company. The formula is Square Roots of (22,5 * Profit per Unit * Equity Value). We are going to keep companies that have intrinsic value greater than stock price.


- PEG Ratio: The PEG ratio enhances the P/E ratio by adding in expected earnings growth into the calculation. Lower PEG Ratios are usually better

```python
def graham_formula(data): 
    data.insert(2,'Graham Formula','')
    data['Graham Formula'] = np.sqrt(22.5 * (data['Cotação']/data['P/L']) * (data['Cotação']/data['P/VP']))
    
def peg_ratio(data):
    data.insert(4, 'PEG. Ratio', '')
    data['PEG. Ratio'] = np.divide(data['P/L'], data['Cresc. Rec.5a']) / 100
```

```python
graham_formula(filtered_df)
peg_ratio(filtered_df)

filtered_df = filtered_df[filtered_df['Graham Formula'] > filtered_df['Cotação']]
```

```python
filtered_df
```

Now, we are keeping companies that have 
- Growth Ratio of the last 5 years greather than 0 
- PEG Ratio < 1 and
- Current assets over current liabilities > 1 (Result that shows available slack for a possible settlement of obligations)

```python
filtered_df = filtered_df[(filtered_df['Cresc. Rec.5a'] > 0) & (filtered_df['PEG. Ratio'] < 1) & (filtered_df['Liq. Corr.'] > 1)]
```

```python
filtered_df
```

Below, we start the second part of the analysis.

We are going to use the Magic Formula of Joel Greenblatt: https://www.investopedia.com/terms/m/magic-formula-investing.asp

```python
# Limitando Liquidez
new_df = dataframe[dataframe['Liq.2meses'] > 800000]
```

```python
# Criando a formula mágica de Joel Greenblatt

ranking = pd.DataFrame({'Posição': range(1,151)})
ranking['EV/EBIT'] = new_df[new_df['EV/EBIT'] > 0].sort_values(by=['EV/EBIT'], ascending=True)['Papel'][:150].values
ranking['ROIC'] = new_df.sort_values(by=['ROIC'], ascending=False)['Papel'][:150].values
```

```python
ranking
```

```python
# Creating pivot table

l1 = ranking.pivot_table(columns='EV/EBIT', values='Posição')
l2 = ranking.pivot_table(columns='ROIC',values='Posição')

pivot_table = pd.concat([l1, l2])
```

```python
pivot_table
```

```python
# Deleting Column that has NaN value
pivot_table.dropna(axis=1)
```

```python
# Summing position values and sorting
magic_ranking = np.sum(pivot_table).sort_values()
```

The correct way to analysis the magic formula is:
- the less points the stock has, the better it is

```python
magic_ranking[:51]
```

Here We compare both datasets (Filtered dataframe and Magic Formula Ranking) finding stocks that are contained in both

The result shall be good options of stock picks for you

```python
# Comparing Magic Ranking and Filtered dataframe to find common tickers

magic_tickers = np.array(magic_ranking[:21].index).transpose()
filtered_tickers = filtered_df['Papel'].values

common_tickers = []
```

```python
for i in range(0, len(magic_tickers)):
    for j in range(0, len(filtered_tickers)):
        if magic_tickers[i] == filtered_tickers[j]:
            common_tickers.append(magic_tickers[i])
```
